```
import pandas as pd
url = 'https://raw.githubusercontent.com/lushanthan/datascience/main/corona_deaths.csv'
df = pd.read_csv(url)

#Display the first 10 rows
result = df.head(10)
print("First 10 rows of the Corona Deaths CSV:")
print(result)

```

```
import matplotlib.pyplot as plt

ax = df.plot.line(x='start_date', y='deaths', grid=True,label="No Of Deaths in the USA", linewidth=2, figsize=(15, 5))
ax.set_xticks(range(50));
ax.set_xticklabels( labels = df["start_date"] ,rotation=90);
```

```
df['deaths_smv_3'] = df.iloc[:,7].rolling(window=3).mean()
df.head(10)
```

```
ax = df.plot.line(x='start_date', y='deaths', grid=True,label="No Of Deaths in the USA", linewidth=2, figsize=(15, 5))
ax.set_xticks(range(50));
ax.set_xticklabels( labels = df["start_date"] ,rotation=90);

plt.plot(df['deaths_smv_3'],label='Simple Moving Average - 3Weeks')
plt.legend(loc=2)
```

```
df['deaths_smv_5'] = df.iloc[:,7].rolling(window=5).mean()
df.head(10)
```

```
ax = df.plot.line(x='start_date', y='deaths', grid=True,label="No Of Deaths in the USA", linewidth=2, figsize=(15, 5))
ax.set_xticks(range(50));
ax.set_xticklabels( labels = df["start_date"] ,rotation=90);

plt.plot(df['deaths_smv_3'],label='Simple Moving Average -  3 Weeks')

plt.plot(df['deaths_smv_5'],label='Simple Moving Average -  5 Weeks')
plt.legend(loc=2)
```
